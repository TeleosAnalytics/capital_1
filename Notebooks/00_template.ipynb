```python
# 1. Anchor to your robust config path
landing_zone = cptl.config.proj_path / "data" / "landing_zone"

# 2. Sweep for the files
svs_files = list(landing_zone.rglob("*.svs"))

# 3. Extract the metadata points
metadata = []
for file_path in svs_files:
    # file_path.parent.name gives you the UUID folder name
    file_id = file_path.parent.name
    
    # file_path.stem strips the .svs extension off the filename
    prefix = file_path.stem 
    
    # TCGA barcodes contain the Patient ID in the first 12 characters
    patient_id = prefix[:12] if prefix.startswith("TCGA") else "Unknown"
    
    # Get physical file size in Megabytes
    size_mb = round(file_path.stat().st_size / (1024 * 1024), 2)

    metadata.append({
        "id": file_id,
        "prefix": prefix,
        "patient_id": patient_id,
        "size_mb": size_mb,
        "path": str(file_path.resolve()) # .resolve() guarantees a strict absolute path
    })

# 4. Forge the DataFrame
metadf = pd.DataFrame(metadata)

# Display the ledger
metadf
```
